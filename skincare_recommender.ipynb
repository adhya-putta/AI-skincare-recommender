## üßº Data Cleaning
Removed missing values, normalized text casing, dropped duplicates.

```python
import pandas as pd

# Load your skincare dataset
df = pd.read_csv("skincare_products.csv")

# Clean casing and whitespace
df['Skin Type'] = df['Skin Type'].str.strip().str.title()
df['Concern'] = df['Concern'].str.strip().str.title()

# Drop missing values and duplicates
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)

# Preview
df.head()
```

## üìä Exploratory Data Analysis
Visualized top concerns, average prices, and product ratings.

```python
import matplotlib.pyplot as plt

df['Concern'].value_counts().plot(kind='bar', color='lightblue')
plt.title('Top Skincare Concerns')
plt.xlabel('Concern')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

## ü§ñ Recommendation Logic
Filtered products by user inputs and returned top-rated matches.

```python
# Simulated user input
user_skin_type = "Oily"
user_concern = "Acne"

# Filter matching products
results = df[
    df["Skin Type"].str.contains(user_skin_type, case=False) &
    df["Concern"].str.contains(user_concern, case=False)
]

# Top 3 recommendations
results.sort_values(by="Rating", ascending=False).head(3)
```

## üìà Insights
Listed takeaways from the visual and numeric analysis.

```python
df['Concern'].value_counts().plot(kind='bar', color='lightblue')
plt.title('Most Common Skincare Concerns')
plt.xlabel('Concern')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
df.groupby('Skin Type')['Price ($)'].mean().sort_values().plot(kind='barh', color='salmon')
plt.title('Avg. Price by Skin Type')
plt.xlabel('Avg. Price ($)')
plt.show()
```

```python
df.plot.scatter(x='Price ($)', y='Rating', color='mediumorchid')
plt.title('Product Price vs. Rating')
plt.show()
```
